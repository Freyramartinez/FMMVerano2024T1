```python
# Importaciones necesarias
import matplotlib.pyplot as plt
import numpy as np

# Funci√≥n para formatear los strings en los datos de entrada
def formatar_strings(datos):
    return [str(item).strip() for item in datos]

# Funci√≥n para ordenar las clases de forma ascendente
def ordenar_clases_asc(datos):
    return sorted(set(datos))

# Funci√≥n para calcular la frecuencia absoluta de los datos
def frecuencia_absoluta(datos, clases):
    fa = [datos.count(clase) for clase in clases]
    return clases, fa

# Funci√≥n para calcular la frecuencia relativa
def frecuencia_relativa(fa):
    total = sum(fa)
    return [round(f / total, 2) for f in fa]

# Funci√≥n para calcular la frecuencia relativa acumulada
def frecuencia_acumulada(frecuencia_rel):
    acumulada = []
    suma = 0
    for fr in frecuencia_rel:
        suma += fr
        acumulada.append(suma)
    return acumulada

# Funci√≥n para imprimir la tabla con los datos ordenados
def imprimir_tabla(clases, fa, frecuencia_rel, frecuencia_rel_acum):
    print("{:<10} {:<10} {:<10} {:<10}".format("Clase", "Frec Abs", "Frec Rel", "Frec Rel Acum"))
    for i in range(len(clases)):
        print("{:<10} {:<10} {:<10} {:<10}".format(clases[i], fa[i], frecuencia_rel[i], frecuencia_rel_acum[i]))

# Funci√≥n para graficar el diagrama de barras
def graficar_barras(clases, fa):
    y_pos = np.arange(len(clases))
    plt.barh(y_pos, fa, align='center', color=['blue', 'red', 'green', 'purple', 'orange'])  # Colores personalizados
    plt.yticks(y_pos, clases)
    plt.ylabel('Clases')
    plt.xlabel('Frecuencia Absoluta')
    plt.title('Diagrama de Barras')
    plt.show()

# Funci√≥n para graficar el diagrama de pastel
def graficar_pastel(fa, clases):
    plt.pie(fa, labels=clases, autopct='%1.1f%%', startangle=140)
    plt.axis('equal')
    plt.title('Diagrama de Pastel')
    plt.show()

# Datos de entrada
datos = [
  "Pumas", "Am√©rica", "Chivas", "Pumas", "Chivas",
  "Chivas", "Am√©rica", "Toluca", "Toluca", "Pumas",
  "Santos", "Necaxa", "Am√©rica", "Am√©rica", "Am√©rica",
  "Pumas", "Santos", "Pumas", "Pumas", "Pumas",
  "Am√©rica", "Chivas", "Am√©rica", "Pumas", "Necaxa",
  "Toluca", "Chivas", "Am√©rica", "Toluca", "Santos"
]

# Formatear los datos
datos_formateados = formatar_strings(datos)

# Ordenar las clases
datos_ordenados = ordenar_clases_asc(datos_formateados)

# Calcular la frecuencia absoluta
clases, fa_absoluta = frecuencia_absoluta(datos_formateados, datos_ordenados)

# Calcular frecuencia relativa
frecuencia_relativa = frecuencia_relativa(fa_absoluta)

# Calcular frecuencia relativa acumulada
frecuencia_acumulada = frecuencia_acumulada(frecuencia_relativa)

# Imprimir tabla
imprimir_tabla(clases, fa_absoluta, frecuencia_relativa, frecuencia_acumulada)

# Graficar diagrama de barras
graficar_barras(clases, fa_absoluta)

# Graficar diagrama de pastel
graficar_pastel(fa_absoluta, clases)
```
