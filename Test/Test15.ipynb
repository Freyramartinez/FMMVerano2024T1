```python
import matplotlib.pyplot as plt

class FormatearStrings:
    @staticmethod
    def format_str(datos):
        return [dato.strip() for dato in datos]

class clas_str_sort:
    @staticmethod
    def clases_str_sort(datos_formated):
        return sorted(datos_formated)

def frec_abs(datos_sorted):
    clases = sorted(set(datos_sorted))
    fa_ab = [datos_sorted.count(clase) for clase in clases]
    return clases, fa_ab

class OrdenarArregloAsendente:
    @staticmethod
    def ordenar_asc(arreglo):
        return sorted(arreglo)

class OrdenarFrecAbsolutas:
    @staticmethod
    def sort_clases_fa(clases_orden, clases_asc, fa_ab):
        fa_sorted = []
        for clase in clases_asc:
            index = clases_orden.index(clase)
            fa_sorted.append(fa_ab[index])
        return fa_sorted

class FrecRelativa:
    @staticmethod
    def frec_rel(fa_sorted):
        total = sum(fa_sorted)
        return [fa / total for fa in fa_sorted]

class FrecAcumulada:
    @staticmethod
    def frec_acum(frecuencia_rel):
        frecuencia_rel_acum = []
        acumulado = 0
        for fr in frecuencia_rel:
            acumulado += fr
            frecuencia_rel_acum.append(acumulado)
        return frecuencia_rel_acum

class Tabla4Datos:
    @staticmethod
    def tabla(clases, fa_sorted, frecuencia_rel, frecuencia_rel_acum):
        print("Clase\tFrecuencia Absoluta\tFrecuencia Relativa\tFrecuencia Relativa Acumulada")
        for i in range(len(clases)):
            print(f"{clases[i]}\t{fa_sorted[i]}\t\t\t{frecuencia_rel[i]}\t\t\t{frecuencia_rel_acum[i]}")

class DiagramaPastel:
    @staticmethod
    def plot_pie(fa_sorted, clases):
        plt.figure(figsize=(8, 8))
        plt.pie(fa_sorted, labels=clases, autopct='%1.1f%%', startangle=140)
        plt.axis('equal')
        plt.title('Diagrama de Pastel de Frecuencias Absolutas')
        plt.show()

# Datos de entrada
datos = ["Ramirez", "Rodrigez", "Hernandez", "Hernandez", "Rodrigez", "Rodrigez", "Gonzalez", "Gonzalez", 
         "Ramirez", "Hernandez", "Garcia", "Martinez", "Lopez", "Gonzalez", "Martinez", "Lopez", "Lopez", 
         "Gonzalez", "Gutierrez", "Martinez", "Gonzalez", "Rodrigez", "Perez", "Rodrigez", "Ramirez", "Sanchez",
         "Martinez", "Lopez", "Sanchez", "Garcia", "Rodrigez", "Hernandez","Gonzalez", "Sanchez", "Perez", 
         "Rodrigez", "Lopez", "Martinez", "Flores", "Hernandez", "Perez", "Garcia", "Martinez", "Hernandez",
         "Sanchez", "Ramirez", "Garcia", "Gonzalez", "Garcia", "Flores", "Martinez", "Garcia", "Ramirez", "Gonzalez",
         "Hernandez", "Lopez", "Sanchez", "Martinez", "Martinez", "Perez", "Garcia", "Gonzalez", "Garcia", "Gutierrez",
         "Hernandez", "Martinez", "Garcia", "Hernandez", "Garcia", "Gonzalez", "Lopez", "Hernandez", "Rodrigez", 
         "Martinez", "Gonzalez", "Garcia", "Hernandez", "Ramirez", "Flores", "Hernandez", "Gutierrez", "Garcia", "Hernandez",
         "Gonzalez", "Garcia", "Garcia", "Flores", "Lopez", "Perez", "Sanchez", "Lopez", "Gutierrez", "Sanchez", "Rodrigez",
         "Hernandez", "Hernandez", "Perez", "Sanchez", "Lopez", "Martinez"]

# Formatea los datos para evitar espacios y excepciones en cadena
datos_formated = FormatearStrings.format_str(datos)

# Ordena la lista de datos
datos_sorted = clas_str_sort.clases_str_sort(datos_formated)

# Calcula la frecuencia absoluta de los datos
clases, fa_ab = frec_abs(datos_sorted)

# Crea una copia de las clases originales sin ordenar
clases_orden = clases.copy()

# Obtiene las frecuencias absolutas ordenadas ascendente
fa_sorted = OrdenarFrecAbsolutas.sort_clases_fa(clases_orden, OrdenarArregloAsendente.ordenar_asc(clases), fa_ab)

# Calcula la frecuencia relativa
frecuencia_rel = FrecRelativa.frec_rel(fa_sorted)

# Calcula la frecuencia relativa acumulada
frecuencia_rel_acum = FrecAcumulada.frec_acum(frecuencia_rel)

# Imprime las clases, su frecuencia relativa y su frecuencia relativa acumulada ordenadas de forma ascendente
Tabla4Datos.tabla(clases, fa_sorted, frecuencia_rel, frecuencia_rel_acum)

# Llama a las gráficas para ingresar sus parámetros
DiagramaPastel.plot_pie(fa_sorted, clases)  # Gráfico de pie
```
